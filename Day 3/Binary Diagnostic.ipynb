```python
# Find the most common bit for each position in the binary output

with open('Binary Output.txt', 'r') as binaryfile:
    binaryData = binaryfile.read()

binary = binaryData.split('\n')

gamma = [None] * 12
epsilon = [None] * 12

# Find the most common bit in a row, bitIndex
for bitIndex in range (0, 12):
    counter0 = 0
    counter1 = 0    
    for index, line in enumerate(binary):
        bit = int(line[bitIndex])
        if(bit == 0):
            counter0 += 1
        if(bit == 1):
            counter1 += 1

    if(counter0 > counter1):
        gamma[bitIndex] = 0
        epsilon[bitIndex] = 1
    elif(counter1 > counter0):
        gamma[bitIndex] = 1
        epsilon[bitIndex] = 0
    else: 
        print('ERROR - Equal amounts of 1 and 0')

strgamma = ""
for cha in gamma:
    strgamma += str(cha)

strepsilon = ""
for cha in epsilon:
    strepsilon += str(cha)

# Convert to base 10
gammaVal = int(strgamma, base=2)
epsilonVal = int(strepsilon, base=2)

print(f'gamma: {gamma}')
print(f'epsilon: {epsilon}')

print(f'strgamma: {strgamma}')
print(f'strepsilon: {strepsilon}')

print(f'gammaVal: {gammaVal}')
print(f'epsilonVal: {epsilonVal}')

powerConsumption = gammaVal * epsilonVal

print(f'power consumption: {powerConsumption}')



```

```python
# Life support rating

# Oxygen generator rating 
# Find the most common value in the current bit position and keep only numbers with that bit in that position.
# If equally common, keep values with a 1.

# Copy the dataset
unculledbinary = binary.copy()

# Find the most common bit in a row, bitIndex
for bitIndex in range (0, 12):

    print(f'unculledbinary length: {len(unculledbinary)}')

    # Find most common bit
    counter0 = 0
    counter1 = 0    
    for index, line in enumerate(unculledbinary):
        bit = int(line[bitIndex])
        if(bit == 0):
            counter0 += 1
        if(bit == 1):
            counter1 += 1

    if(counter0 > counter1):
        mostcommon = 0
    elif(counter1 > counter0):
        mostcommon = 1
    else: 
        # Defaulting to 1 will mean that when equal amounts, those with 1 will be kept
        mostcommon = 1

    print(f'most common: {mostcommon}')

    # Cull the data
    # Create empty array to store surviving values
    culledbinary = []
    for index, line in enumerate(unculledbinary):
        if(int(line[bitIndex]) != mostcommon):
            pass
        elif(int(line[bitIndex]) == mostcommon):
            # Save to next iteration
            culledbinary.append(line)
            
    # If only one left, then end the culling
    if(len(culledbinary) == 1):
        break

    unculledbinary = culledbinary.copy()

print(culledbinary)
oxygenrating = culledbinary[0]
```
