```python
import numpy as np
```

```python
with open('Boards.txt') as boardfile:
    boardDataRaw = boardfile.read()

boardData = boardDataRaw.split('\n\n')
#print(len(boardData))
numbers = boardData[0]
boardblocks = boardData[1:len(boardData)]
#print(len(boardblocks))

# Construct 2D arrays for the boards
boards = np.array([])
# For each board in boards, construct a 2D array for its values
for i, boardblock in enumerate(boardblocks):
    print(i)
    # Create a board array
    board = np.array([])
    # Get the data for the rows
    rows = boardblock.split('\n')
    # Iterate over the rows, extracting each row
    for j, row in enumerate(rows): 
        # For each row, add an array to board
        board = np.append(board, [''])
        values = row.split(' ')
        # For each value in the row
        for k, value in enumerate(values):
            # Append the value to the row
            print(value)
            np.append(board[j], value)

    # Add the board to the array of boards
    boards = np.append(boards, board)

for i in range(0, 4):
    print(boards[i])



# # Iterate through the numbers
# for index, value in enumerate(numbers):
#     # Check boards for the number#
#     for 
```
