```python
import math
with open('input.txt') as file:
    lines = file.read().split('\n')
# print(lines)

invalid_list = []

opening_brackets = ['(', '{', '[', '<']
closing_brackets = [')', '}', ']', '>']
bracket_pairs = ['()', '{}', '[]', '<>']

closing_char_scores = {
    ')': 1,
    ']': 2,
    '}': 3,
    '>': 4,
}

filtered_lines = []

for line_index, line in enumerate(lines):
    # Iteratively find closed brackets
    line = list(line)
    i = 0
    j = 0
    # print(line)
    while j < len(line) - 1:
        j = i + 1
        substr = line[i] + line[j]
        # print(f'  substr: {substr}')
        if substr in bracket_pairs:
            # print(f'  match. removing {i} and {j}')
            line.pop(i)
            # pop i twice, because now j is i
            line.pop(i) 
            # print(line)
            i = 0
            j = 1
        elif substr[0] in opening_brackets and substr[1] in closing_brackets:
            # invalid!
            invalid_list.append(line_index)
            break
        else:
            i += 1
    else:
        filtered_lines.append(line)
    # print(line)
    # print('')
    # print('')          

invalid_list.sort(reverse=True)
print(invalid_list)  
for i in invalid_list:
    lines.pop(i)

print('filtered lines: ')
for line in filtered_lines:
    print(line)
    print('')
print('-------------')
# We now have either only valid lines
closing_scores = []
for line in filtered_lines: 
    closing_score = 0
    print(line)
    print('')
    closing_string = ''
    while len(line) > 0:
        last_char = line[len(line)-1]
        char_to_add = closing_brackets[opening_brackets.index(last_char)]
        closing_string += char_to_add
        closing_score *= 5
        closing_score += closing_char_scores[char_to_add]
        line.pop(len(line)-1)
    print(closing_string)
    closing_scores.append(closing_score)
    print(f'score: {closing_score}')
    print(' ')
    print('--------')

print('')
closing_scores.sort()
print(f'closing score: {closing_scores}')
print(f'len closing scores: {len(closing_scores)}')
print(f'closing score overall: {closing_scores[math.floor(len(closing_scores)/2)]}')
```
