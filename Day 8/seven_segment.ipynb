```python
# with open('./test_input.txt') as file:
#     lines = file.read().split('\n')
# input_outputs = [line.split(' | ') for line in lines]
# print(f'input/outputs: {input_outputs}')

total = 0
with open('./input.txt') as file:
    lines = file.read().split('\n')

for input, output in [line.split(' | ') for line in lines]:
    # print(f'input: {input}')
    # print(f'output: {output}')
    # print(f'length_dict: ')
    length_dict = {len(signal): set(signal) for signal in input.split(' ')}
    # print(length_dict)

    n = ''
    # Convert output to a set 
    for out in map(set, output.split()): 
        # print(f' out: {out}')
        # This matches the length of the output, the length of the output set along with the 4th
        # print(f'  length_dict[4]: {length_dict[4]}')
        # print(f'  out&length_dict[4]: {out&length_dict[4]}')
        # print(f'  len(out): {len(out)}')
        # print(f'  len(out&l[4]): {len(out&length_dict[4])}')
        # print(f'  len(out&l[2]: {len(out&length_dict[2])}')
        # One, Seven, Four and Eight have unique numbers of segments and so are immediately identifiable.
        # Two, Five and Three have 5 segments lit, therefore we need to compare some more components.
            # We can use the already known numbers to compare the number of segments that are in common, to determine further values.
            # Two and Four have 2 segments in common, Five and Three have 3 segments in common, therefore we can find Two.
            # Five and Three have 3 segments in common with Four, therefore we need another known value. Turns out we can use One, whose length is 2.
        # Repeat this process for the rest of the outputs
        match len(out), len(out&length_dict[4]), len(out&length_dict[2]):
            case 2,_,_: n += '1'
            case 3,_,_: n += '7'
            case 4,_,_: n += '4'
            case 7,_,_: n += '8'
            case 5,2,_: n += '2'
            case 5,3,1: n += '5'
            case 5,3,2: n += '3'
            case 6,4,_: n += '9'
            case 6,3,1: n += '6'
            case 6,3,2: n += '0'
    total += int(n)
    print(f' {total}')
print(total)




```
